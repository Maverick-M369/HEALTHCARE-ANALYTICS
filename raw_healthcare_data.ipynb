```python
import pandas as pd
import numpy as np

np.random.seed(42)

ROWS = 1200  # >1000 rows

patients = [f"P{str(i).zfill(4)}" for i in range(1, 501)]
departments = ["Cardiology", "Orthopedics", "ICU", "General", "Neurology"]

bed_capacity = {
    "Cardiology": 120,
    "Orthopedics": 90,
    "ICU": 50,
    "General": 180,
    "Neurology": 70
}

data = {
    "patient_id": np.random.choice(patients, ROWS),
    "admission_id": [f"A{10000+i}" for i in range(ROWS)],
    "department": np.random.choice(departments, ROWS),
}

admission_dates = pd.to_datetime("2023-01-01") + \
                  pd.to_timedelta(np.random.randint(0, 365, ROWS), unit="D")

los = np.random.randint(1, 16, ROWS)
discharge_dates = admission_dates + pd.to_timedelta(los, unit="D")

data["admission_date"] = admission_dates
data["discharge_date"] = discharge_dates

data["readmitted"] = np.random.choice(
    ["Yes", "No"], size=ROWS, p=[0.27, 0.73]
)

data["total_beds"] = [bed_capacity[d] for d in data["department"]]
data["occupied_beds"] = [
    np.random.randint(int(tb*0.6), tb+1) for tb in data["total_beds"]
]

df_raw = pd.DataFrame(data)

df_raw.to_csv("M:/Data Science/Data Analyst/Github & LinkedIN Portfolio Projects/Healthcare Analytics/hospital_admissions_raw.csv", index=False)

df_raw.head()
```
